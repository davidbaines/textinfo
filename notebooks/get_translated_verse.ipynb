```python
# import the library
from itertools import groupby
from operator import itemgetter
from pathlib import Path
import re
import yaml

config_filename = "config.yml"
test_vref_filename = "test.vref.txt"
test_src_filename = "test.src.txt"
test_trg_filename = "test.trg.detok.txt"

greek_source = "grc-grcsbl"

def get_lines(file):
    with open(file, "r", encoding='utf-8') as f:
        return [line.strip() for line in f.readlines()]
 
def strip_lang_code(line: str) -> str:
    return re.sub(r" ?</s> ?[a-zA-Z]{3}_[a-zA-Z]{4}$", "", line)

def decode_sp(line: str) -> str:
    #    return line.replace(" ", "").replace("\u2581", " ").lstrip()
    # Unicode 2581 (Lower One Eighth Block) used by SentencePiece for start-of-word indicator
    # Unicode 2580 (Upper Half Block) used for morpheme segment indicator
    return line.replace(" ", "").replace("\u2581\u2580", "").replace("\u2581", " ").lstrip()


def detok_line(line):
    return decode_sp(strip_lang_code(line))


vrefs_file = Path("F:/GitHub/silnlp/silnlp/assets/vref.txt")


with open(vrefs_file, encoding='utf-8') as f_vrefs:
    vrefs = {idx: line.strip() for idx, line in enumerate(f_vrefs.readlines())}
#print(vrefs.values())

folder = Path("F:\GitHub\BibleNLP\ebible-experiments\MT\experiments")


# Get the experiments details from the config.yml file.
experiments = []

test_vref_files = [file for file in folder.rglob(test_vref_filename)]

for test_vref_file in test_vref_files:
    lines = get_lines(test_vref_file)
    if "JHN 1:1" in lines:
        folder = test_vref_file.parent
        line_index = lines.index("JHN 1:1")
        experiment = {
            "test_vref_file": test_vref_file,
            "line_index": line_index,
            "folder": folder,
            "config_file" : folder / config_filename,
            "test_src_file" : folder / test_src_filename,
            "test_trg_file" : folder /test_trg_filename,

            }
        
        with open(experiment["config_file"], "r") as configfile:
            config = yaml.safe_load(configfile)
        
        experiment["config"] = config

        experiment["src_line"] = detok_line(get_lines(experiment["test_src_file"])[line_index])
        experiment["trg_line"] = get_lines(experiment["test_trg_file"])[line_index]
        print(experiment["folder"])
        print(experiment["src_line"])
        print(experiment["trg_line"])
        print()


        experiments.append(experiment)
    
```
