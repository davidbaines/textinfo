```python
from pathlib import Path
from pprint import pprint
import shutil
from typing import Dict
import yaml

experiments_folder_str = "S:/MT/experiments"
experiments_folder = Path(experiments_folder_str)
destination_folder = Path("C:/Gutenberg/MT/experiments/")

family_folder_prefix = 'eBible.NMT.'
source_language_family = 'Niger-Congo'
source_family_folder = experiments_folder / (family_folder_prefix + source_language_family)


destination_language_families = ['Afro-Asiatic','Austronesian', 'Dravidian', 'Indo-European', "Niger-Congo", "Otomanguean", "Sino-Tibetan", "Trans-NewGuinea",]

# Set the series to create:
#destination_language_family = 'Afro-Asiatic'
destination_language_family = 'Indo-European'

destination_family_folder = destination_folder / (family_folder_prefix + destination_language_family)
print(f"The source_folder is {source_family_folder}\nThe destination is   {destination_family_folder}")
#pprint(language_details)


file_patterns_to_copy = ['config.yml']
```

```python
source_subfolders = [folder for folder in source_family_folder.iterdir()]
print(f"Found {len(source_subfolders)} source subfolders")
for source_subfolder in source_subfolders:
    print(source_subfolder)

```

```python
experiments = []

print(f"Looking for files to copy.")
for source_subfolder in source_subfolders:

    #print(f"Source is      : {source_subfolder}")
    for file_pattern_to_copy in file_patterns_to_copy:
        for source_file in source_subfolder.rglob(file_pattern_to_copy):

            if source_file.is_file():
                    
                dest_folder = destination_family_folder / source_subfolder.name            
                destination_file = dest_folder / source_file.name

                # Get the experiment details from that config.yml file.
                with open(source_file, "r") as configfile:
                    config: Dict = yaml.safe_load(configfile)

                experiments.append({"Source folder": source_subfolder, "Source file": source_file,
                                      "Destination file" : destination_file, "Destination folder" : dest_folder,
                                      "Config": config, "Language family": destination_language_family})

print(f"\nExperiment details are:")
pprint(experiments[0])
```

```python
for experiment in experiments:
    print(experiment['Destination file'])
    #Write the same output:
    if not experiment['Destination folder'].is_dir():
        experiment['Destination folder'].mkdir(parents=True, exist_ok=True)
        print(f"Created destination folder: {dest_folder}")
        
    with open(experiment['Destination file'], "w") as configfile:
        yaml.dump(experiment['Config'], configfile, default_flow_style=False)
```
